# Lets load the deep learning model

```python
mobile = tf.keras.applications.mobilenet.MobileNet() #Deep learning model weights pre-trained
```

```python
# Loading using MobileNetV2
```

```python
mobile = tf.keras.applications.mobilenet_v2.MobileNetV2()
```

# Pre-processing the image

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename, target_size = (224,224))
```

```python
plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img,axis = 0) # Creating fourth dimension
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
predictions = mobile.predict(final_image)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```

```python
predictions = mobile.predict(final_image) # predicting using MobileNetV2
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
print(results) # MobileNetV2 is more accurate
```
